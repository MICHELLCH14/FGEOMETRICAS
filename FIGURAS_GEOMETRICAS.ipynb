```python
import math
import datetime
from abc import ABC, abstractmethod


class Color:
    def __init__(self, color):
        self._color = None
        self.color = color

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, value):
        if isinstance(value, str) and value.strip():
            self._color = value
        else:
            raise ValueError("El color debe ser una cadena de texto no vacía.")

    def __str__(self):
        return f"Color: {self.color}"


class FiguraGeometrica(ABC):
    def __init__(self, alto, ancho):
        self._alto = None
        self._ancho = None
        self.alto = alto
        self.ancho = ancho

    @property
    def alto(self):
        return self._alto

    @alto.setter
    def alto(self, value):
        if isinstance(value, (int, float)) and value > 0:
            self._alto = float(value)
        else:
            raise ValueError("El alto debe ser un número positivo.")

    @property
    def ancho(self):
        return self._ancho

    @ancho.setter
    def ancho(self, value):
        if isinstance(value, (int, float)) and value > 0:
            self._ancho = float(value)
        else:
            raise ValueError("El ancho debe ser un número positivo.")

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimetro(self):
        pass

    def __str__(self):
        return f"Alto: {self.alto}, Ancho: {self.ancho}"


class Rectangulo(FiguraGeometrica, Color):
    def __init__(self, alto, ancho, color):
        FiguraGeometrica.__init__(self, alto, ancho)
        Color.__init__(self, color)

    def area(self):
        return self.alto * self.ancho

    def perimetro(self):
        return 2 * (self.alto + self.ancho)

    def __str__(self):
        return f"Rectángulo -> {FiguraGeometrica.__str__(self)}, " \
               f"{Color.__str__(self)}"


class Cuadrado(FiguraGeometrica, Color):
    def __init__(self, lado, color):
        FiguraGeometrica.__init__(self, lado, lado)
        Color.__init__(self, color)

    @property
    def alto(self):
        return self._alto

    @alto.setter
    def alto(self, value):
        if isinstance(value, (int, float)) and value > 0:
            v = float(value)
            self._alto = v
            self._ancho = v
        else:
            raise ValueError("El lado del cuadrado debe ser un número "
                             "positivo.")

    @property
    def ancho(self):
        return self._ancho

    @ancho.setter
    def ancho(self, value):
        if isinstance(value, (int, float)) and value > 0:
            v = float(value)
            self._alto = v
            self._ancho = v
        else:
            raise ValueError("El lado del cuadrado debe ser un número "
                             "positivo.")

    def area(self):
        return self.alto * self.alto

    def perimetro(self):
        return 4 * self.alto

    def __str__(self):
        return f"Cuadrado -> Lado: {self.alto}, {Color.__str__(self)}"


class Circunferencia(FiguraGeometrica, Color):
    def __init__(self, diametro, color):
        FiguraGeometrica.__init__(self, diametro, diametro)
        Color.__init__(self, color)

    @property
    def alto(self):
        return self._alto

    @alto.setter
    def alto(self, value):
        if isinstance(value, (int, float)) and value > 0:
            v = float(value)
            self._alto = v
            self._ancho = v
        else:
            raise ValueError("El diámetro de la circunferencia debe ser un "
                             "número positivo.")

    @property
    def ancho(self):
        return self._ancho

    @ancho.setter
    def ancho(self, value):
        if isinstance(value, (int, float)) and value > 0:
            v = float(value)
            self._alto = v
            self._ancho = v
        else:
            raise ValueError("El diámetro de la circunferencia debe ser un "
                             "número positivo.")

    def area(self):
        radio = self.alto / 2.0
        return math.pi * (radio ** 2)

    def perimetro(self):
        return math.pi * self.alto

    def __str__(self):
        return f"Circunferencia -> Diámetro: {self.alto}, " \
               f"{Color.__str__(self)}"


if __name__ == "__main__":
    now = datetime.datetime.now()
    fecha_hora_str = now.strftime('%Y-%m-%d %H:%M:%S')
    print(f"--- Ejecución del Código - {fecha_hora_str} ---\n")

    figuras = []
    try:
        figuras.append(Rectangulo(alto=10, ancho=5, color="Rojo"))
        figuras.append(Cuadrado(lado=7, color="Azul"))
        figuras.append(Circunferencia(diametro=4, color="Verde"))
    except ValueError as e:
        print(f"Error al inicializar figuras: {e}")

    print("--- Demostración de Polimorfismo ---")
    for figura in figuras:
        print(f"\nProcesando figura: {figura}")
        print(f"  Área: {figura.area():.2f}")
        print(f"  Perímetro: {figura.perimetro():.2f}")

    print("\n--- Demostración de Validaciones ---")
    print("Captura 1: Creación de un rectángulo con ancho negativo.")
    try:
        Rectangulo(10, -5, "Rojo")
    except ValueError as e:
        print(f"  RESULTADO -> ERROR: {e}")

    print("\nCaptura 2: Establecer un color inválido (cadena vacía) en un "
          "cuadrado existente.")
    try:
        cuad = figuras[1]
        print(f"  Original: {cuad}")
        cuad.color = " "
    except ValueError as e:
        print(f"  RESULTADO -> ERROR: {e}")
    else:
        print("  Error: La validación del color falló.")

    print("\nCaptura 3: Establecer un lado negativo en un cuadrado "
          "existente.")
    try:
        cuad = figuras[1]
        print(f"  Original: {cuad}")
        cuad.alto = -10
    except ValueError as e:
        print(f"  RESULTADO -> ERROR: {e}")
    else:
        print("  Error: La validación de la dimensión falló.")

    print(f"\nCaptura Final: Verificación de estado de objeto original -> "
          f"{cuad}")
    print(f"\n--- Fin de la Demostración - {fecha_hora_str} ---")
```
